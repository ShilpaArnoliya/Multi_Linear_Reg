```python
# MODEL BUILDING
import statsmodels.formula.api as smf
model = smf.ols('Price ~ Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=ntoyota).fit()
```

```python
model.params
```

```python
model.summary()
# cc and Doors doesnot have much effect on our model as they both hv high p values
#cc p value = 0.179 >0.05
#Doors p value = 0.968 >0.05
# but they dont have any colllinearity in b/w
```

```python
# PREDICTED VALUES
pred = model.predict(ntoyota[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]])
pred
```

```python
#RESIDUALS
resid = pred - ntoyota.Price
resid
```

```python
# RMSE
rmse = np.sqrt(np.mean(resid*resid))
rmse
```

```python
# Observed values(Y) VS Fitted values
plt.scatter(ntoyota.Price,pred ,c="b")
plt.xlabel("observed_values")
plt.ylabel("fitted_values")
plt.show()
```

```python
# Residuals VS Fitted Values 
plt.scatter(pred,model.resid_pearson,c="b")
plt.axhline(y=0,color='blue')
plt.xlabel("fitted_values")
plt.ylabel("residuals")
plt.show()
```

```python
# cc & Price model
import statsmodels.formula.api as smf
model_cc = smf.ols('Price ~ cc',data=ntoyota).fit()
model_cc.summary()

# R-squared: 0.016
```

```python
# Doors & Price model
import statsmodels.formula.api as smf
model_Doors = smf.ols('Price ~ Doors',data=ntoyota).fit()
model_Doors.summary()

# R-squared: 0.034
```

```python
#Influence plot Of MODEL
import statsmodels.api as sm
fig, ax = plt.subplots(figsize=(15, 20))
fig = sm.graphics.influence_plot(model, ax=ax,size=50)
```

```python
# #look like obsv 80,109,110,111 is showing high influence so we can exclude that entire row
data_new=ntoyota.drop(ntoyota.index[[80,109,110,111]],axis=0) #axis=0 remove row , axis=1 is used remove column 
```

```python
# MODEL BUILDING WITH REMOVED INFLUENCED OBSEFVATION
# Dataset = data_new
import statsmodels.formula.api as smf
model1 = smf.ols('Price ~ Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_new).fit()
```

```python
model1.params
```

```python
model1.summary()
#cc p value = 0.179 >0.05   => Now = 0.000
#Doors p value = 0.968 >0.05   => Now  =0.867 
```

```python
# PREDICTED VALUES
pred1 = model1.predict(data_new[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]])
pred1
```

```python
#RESIDUALS
resid1 = pred1 - data_new.Price
resid1
```

```python
# RMSE
rmse1 = np.sqrt(np.mean(resid1*resid1))
rmse1
```

```python
#Influence plot of Deleted influence observations
import statsmodels.api as sm
fig, ax = plt.subplots(figsize=(15, 20))
fig = sm.graphics.influence_plot(model1, ax=ax,size=50)
```

```python
# #look like obsv 960,221,601 is showing high influence so we can exclude that entire row
data_new1=data_new.drop(data_new.index[[960,221,601]],axis=0) #axis=0 remove row , axis=1 is used remove column 
```

```python
# MODEL BUILDING WITH REMOVED INFLUENCED OBSEFVATION
# Dataset = data_new1
import statsmodels.formula.api as smf
model2 = smf.ols('Price ~ Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_new1).fit()
model2.summary()

#there is no cange in the p value of doores
#Doors p value = 0.968 >0.05   => model1  =0.867   => model2   = 0.870
#droping observations in data_new1 results increase in p value so we will not use this dataset
#instead we used unremoved obs. dataset i.e data_new 
#we will do tranformations in model now
```

```python
# LOG TRANSFORMATION Y~LOG(X) 
model_log = smf.ols('Price ~ np.log(Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight)',data=data_new).fit()
model_log.summary()
```

```python
# EXPONENTIAL TRANSFORMATION   log(Y)~x

model_exp = smf.ols('np.log(Price) ~ Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_new1).fit()
model_exp.summary()
```

```python
# QUADRATIC TRANSFORMATION
model_quad = smf.ols('Price ~ (Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight)*(Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight)',data=data_new1).fit()
model_quad.summary()
```

```python
# PREDICTED VALUES
pred_quad = model_quad.predict(data_new1[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]])
pred_quad
```

```python
#RESIDUALS
resid_quad = pred_quad - ntoyota.Price
resid_quad
```

```python
# RMSE
rmse_quad = np.sqrt(np.mean(resid_quad*resid_quad))
rmse_quad
```

```python
# Observed values(Y) VS Fitted values
plt.scatter(data_new1.Price,pred_quad ,c="g")
plt.xlabel("observed_values")
plt.ylabel("fitted_values")
plt.show()
```

```python
# Residuals VS Fitted Values 
plt.scatter(pred_quad ,model_quad.resid_pearson,c="g")
plt.axhline(y=0,color='blue')
plt.xlabel("fitted_values")
plt.ylabel("residuals")
plt.show()
```
